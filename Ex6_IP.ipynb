```python
import cv2
from google.colab.patches import cv2_imshow

img = cv2.imread("wrapping.png")

if img is None:
    print("Image not found")
    exit()

cv2_imshow(img)
cv2.waitKey(0)
cv2.destroyAllWindows()
```

```python
import numpy as np

src_pts = np.float32([
    [420, 170],   # top-left
    [650, 150],   # top-right
    [680, 420],   # bottom-right
    [400, 440]    # bottom-left
])

print("Source Points:\n", src_pts)
```

```python
width = 300
height = 420

dst_pts = np.float32([
    [0, 0],
    [width, 0],
    [width, height],
    [0, height]
])

print("Destination Points:\n", dst_pts)
```

```python
import cv2

H = cv2.getPerspectiveTransform(src_pts, dst_pts)

print("Homography Matrix:\n", H)
```

```python
warped = cv2.warpPerspective(img, H, (width, height))

cv2_imshow(warped)
cv2.waitKey(0)
cv2.destroyAllWindows()
```
